```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import squarify
```

a. crei un dataframe contenente le informazioni riportate nella tabella sottostante, relativa al numero di
ordini (order_numbers) seguiti da vari commessi (last_name) di un negozio.

```python
d = {"last_name" : ["Boyer","Copeland","Daniel","Serrano","Terrel","Vargas"],
      "order_numbers" : [185,59,196,58,24,28]}
df = pd.DataFrame(d)
df
```

esercizio b

```python
fig = plt.figure(figsize= (28,20))
ax = plt.axes()
ax.bar(df["last_name"],df['order_numbers'])

```

c. visualizzi lo stesso grafico ma con altri colori delle barre, un titolo appropriato al grafico e i nomi degli
assi.

```python
fig = plt.figure(figsize= (20,20))
ax = plt.axes()
ax.bar(df["last_name"],df['order_numbers'], color = "black")
ax.set_xlabel('last_name')
ax.set_ylabel('Order numbers')
fig.suptitle('ISTOGRAMMA')
plt.show()
```

d. Visualizzare lo stesso grafico ma con le barre in ordine decrescente

```python
fig = plt.figure(figsize= (8,10))
ax = plt.axes()
ax.bar(df.sort_values("order_numbers",ascending = False)["last_name"], df["order_numbers"].sort_values(ascending = False),color = "c"

ax.set_ylabel('a',fontsize = 20)
ax.set_ylabel('a',fontsize = 20)
fig.suptitle('ISTOGRAMMA',fontsize = 53)
plt.show()
```

e. visualizzi un grafico a barre orizzontali per rappresentare i dat. Il risultato deve essere uguale a questo:

```python
fig = plt.figure(figsize= (8,10))
ax = plt.axes()
ax.barh(df["last_name"].sort_values(ascending = False),df['order_numbers'].sort_values(ascending = False), color = "black")
ax.set_xlabel('Order_numbers',fontsize = 20)
ax.set_ylabel('last_name',fontsize = 20)
fig.suptitle('ISTOGRAMMA',fontsize = 53)
plt.show()

```

f. crei un nuovo dataframe contenente la percentuale di ordini consegnati seguiti da ogni componente
dello staff. Visualizzare il cognome del componente dello staff e la percentuale di ordini. Ordinare in
ordine alfabetico sul cognome del componente dello staff. Il risultato deve essere questo:

```python
d = {"last_name" : ["Copeland","Serrano","Boyer","Daniel","Vargas","Terrell"],
      "order_perc" : [9.242144,10.720887,34.195933,36.229205,5.175601,4.436229]}
df = pd.DataFrame(d)
df
```

g. visualizzi un grafico a torta per rappresentare i dati estratti. Il risultato deve essere questo:

```python
names = list(df['last_name'])
values = list(df['order_perc'])
plt.rcParams.update({'font.size': 25}) #grandezza testo
fig = plt.figure(figsize=[7,7]) ##grandeza torta
ax = plt.axes()
ax.pie(values, labels=names,autopct='%.2f')
plt.show()
```

h. visualizzi lo stesso grafico ma con altri colori delle percentuali e un titolo appropriato al grafico

```python
names = list(df['last_name'])
values = list(df['order_perc'])
colors = ['c','m','r','g',"b","y"]
plt.rcParams.update({'font.size': 25}) #grandezza testo
fig = plt.figure(figsize=[7,7]) ##grandeza torta
ax = plt.axes()
ax.pie(values, labels=names,autopct='%.2f', colors = colors)
fig.suptitle("Grafico a Torta",color = "r")
plt.show()
```

i. rappresentare le stesse informazioni utilizzando come tipo di grafico il TreeMap (google it…)

```python
alfa = list(df["last_name"])
beta = list(df["order_perc"])
squarify.plot(sizes=beta ,label=alfa , alpha=0.9 )
plt.axis("off")
plt.show()
```
