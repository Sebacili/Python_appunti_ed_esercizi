```python
import matplotlib.pyplot as plt
import pandas as pd
```

a. creare un dataframe contenente le informazioni riportate qui sotto, relative a quanti prodotti ha
venduto (‘quantity’) e quanto ha venduto in termini di $ (‘total_amount’) ogni membro dello staff
(colonna “last_name”):

```python
d = {"last_name" : ["Boyer","Copeland","Daniel","Serrano","Terrel","Vargas"],
      "quantify" : [766,213,812,269,100,114],
     "total_amount" : [643049.37,203549.39,744259.20,253183.68,93851.95,123156.90] }
df = pd.DataFrame(d)
df
```

b. visualizzi due grafici a barre affiancati che indichino quantità vendute e valore venduto per ogni
dipendente. Il risultato deve essere questo:

```python
fig, (ax,ax2) = plt.subplots(1,2)    # 2 grafici affiancati
fig.set_size_inches(15,8)
ax.bar(df["last_name"],df['quantify']) #primo grafico
ax2.bar(df["last_name"],df['total_amount']) #secondo grafico
plt.show()


```

c. visualizzi i grafici precedenti con un titolo, gli assi e la griglia.

```python
fig, (ax,ax2) = plt.subplots(1,2)    # 2 grafici affiancati
fig.set_size_inches(15,8)
ax.bar(df["last_name"],df['quantify']) #primo grafico
ax2.bar(df["last_name"],df['total_amount']) #secondo grafico
fig.suptitle("istogrammi affiancati",fontsize = 25)
ax.set_ylabel('quantity',fontsize = 15)
ax.set_xlabel('last_name',fontsize = 15)
ax2.set_ylabel('total_amount',fontsize = 15)
ax2.set_xlabel('last_name',fontsize = 15)
ax.grid(linewidth=0.5,color = "k", linestyle = "-")
ax2.grid(linewidth=0.5,color = "k", linestyle = "-")
plt.show()
```

d. visualizzi, senza creare nuovi dataframe, il costo medio dei prodotti venduti da ogni dipendente. Il
risultato deve essere questo (prestare attenzione ai valori dell’asse delle ordinate):

```python
media = list(df["total_amount"] / df["quantify"])
fig = plt.figure(figsize = (10,8))
ax = plt.axes()
ax.bar(df["last_name"],media)
plt.show()
```
